```python
def code(m, k):
    indice=0
    for i in range(k-1,-1,-1):
        indice+=m[k-1-i]*(4**i)
    return indice 
m = [3,0,1]
print(code(m,3))
```

```python
def decode(i,k):
    if i==0:
        return [0]*k
    else:
        mot=[]
        for j in range(k):
            mot.append(i%4)
            i = i//4
        mot = list(reversed(mot))
        return mot
print(decode(49,3))
        
```

```python
def nbOccurencesMot(sequences,k):
    nbOcc = [0 for i in range(4**k)]
    for i in range(len(sequences)-k+1):
        mot = [sequences[i+j] for j in range(k)] # on prend le mot de longeur k à partir du i ieme caractère de la sequence
        indice = code(mot,k)
        nbOcc[indice]+=1
    return nbOcc
seq = [0,3,1,0,3]
print(nbOccurencesMot(seq,2))
```

### Question 2
En supposant que les lettres $(A,T,C,G)$ apparaissent de manière indépendante ,le nombre attendu d’occurrences $nb$ pour un mot $w$ de longeur $k$ dans une séquence de longueur $l$ peut se calculer de la manière suivante:
On calcul le nombre de mots possibles de longeur $k$, ($l-k+1$), qu'on multiplie par la probabilité d'obtenir le mot $w$. On calcul cette probablité en utilisant les fonctions écrites précedemment.
$$ nb = (l-k+1)\cdot P(w)$$

```python

def comptage_attendu(f,k,l):
    nbOcc = []
    n= l-k+1
    for i in range(4**k):
        m = decode(i,k)
        count = [m.count(i) for i in range(4)]
        p = exp(logprobafast(count,f))
        nbOcc.append(n*p)
    return nbOcc
print(comptage_attendu(frequence,2,len(seq)))
```

```python
def simule_sequence(lg, m): 
    """ entrée: lg (longueur de la séquence) , m (fréquence de chaque lettre dans la séquence)
        sortie : séquence aléatoirement simulée dont chaque lettre possède les mêmes fréquences d'apparition que dans m 
    """
    seq = []
    for i in range(lg):
        r = random.random()
        if r<m[0]:
            seq.append(0)
        elif r>=m[0] and r<(m[1]+m[0]):
            seq.append(1)
        elif r>=(m[1]+m[0]) and r<(m[0]+m[1]+m[2]):
            seq.append(2)
        else:
            seq.append(3)
    return seq
```

```python
def simulationComparaison(lg,m,nb,k):
    """entrée: lg (longueur de la séquence) , m (fréquence d'apparition des lettres dans la séquence) , 
             : nb (nombre de silmulations à effectuer), k (la longeur d'un mot donné )
       sortie: graphe comparatif entre le nombre de mots de longeur k attendu et le nombre de mots de longeur k observé
    """
    observe = np.zeros(4**k)
    attendu = np.zeros(4**k)
    for i in range(nb):
        seq = simule_sequence(lg,m)
        ob = np.array(nbOccurencesMot(seq,k))
        att = np.array(comptage_attendu(m,k,lg))
        observe+=ob
        attendu+=att
    x = np.linspace(min(attendu),max(attendu),10)
    plt.plot(x,x, color = 'red')
    plt.scatter(attendu,observe)
    plt.title("pour le mots de taille "+str(k))
    plt.xlabel("nombre attendu")
    plt.ylabel("nombre observé")
simulationComparaison(1000,frequence,100,8)
```

```python
def logProbaEmpiriqueMot(lg,nbTests,liste_mot,k,frequence,n):
    """entrée : lg :longueur de la séquence, nbTests : nombre de tests à effectuer, liste_mots : liste de mots
    sur lequelle on effectuera le test , k : longeur du mot donné , fréquence: fréquence d'apparition de chaque lettre dans la séquence,
    n: le nombre d'apparition du mot donné qu'on veut observé
        sortie: dictionnaire[code du mot de longueur k: probabilité d'avoir une apparition >=n]
    """
    codes = [code(m,k) for m in liste_mot]
    dic = {}
    for c in codes:
        dic[c]=0
        
    for i in range(nbTests):
        seq = simule_sequence(lg,frequence)
        nbOcc = nbOccurencesMot(seq,k)
        for c in codes:
              if nbOcc[c]>=n:
                    dic[c]+=1
    for c in codes:
        dic[c]/=nbTests
    return dic 

liste_mot = [ [0,0,0,0,0,0], [0,3,1,3,2,1], [0,3,0,3,0,3], [3,3,3,0,0,0]]
nbTests = 100
k = 6
l=1000
n=3
logProbaEmpiriqueMot(l,nbTests,liste_mot,k,frequence,n)
```

#### La méthode:
On simule aléatoirement nbTests fois une sequence de longueur lg , on calcule le nombre d'occurences de chaque mot de longueur k , si nbOcc est supérieur ou égale à n le nombre de cas favorable est incrementé de 1 .On divise à la fin par le nombre de tests effetcués .

```python
def estimateM(sequences,frequence):
    M = np.zeros(16)
    for key in sequences:
        seq = openFasta('regulatory_seq_'+key+'.fasta')
        nbOcc = np.array(nbOccurencesMot(seq,2))
        M+= nbOcc
    M = M.reshape((4,4))
    n = M.sum(axis = 1)
    for i in range(4):
        if n[i]!=0:
            M[i,:] = M[i,:]/n[i]
    return M
M = estimateM(sequences,frequence)
print(M)
```

```python
def simulationSequenceDinucleo(l,M,f):
    r = random.random()
    seq = []
    if r<f[0]:
        seq.append(0)
    elif r>=f[0] and r<(f[0]+f[1]):
        seq.append(1)
    elif r>=(f[0]+f[1]) and r<(f[0]+f[1]+f[2]):
        seq.append(2)
    else:
        seq.append(3)
        
    for i in range(1,l):
        r = random.random()
        p = M[seq[i-1]]
        if r<p[0]:
            seq.append(0)
        elif r>=p[0] and r<(p[0]+p[1]):
            seq.append(1)
        elif r>=(p[0]+p[1]) and r<(p[0]+p[1]+p[2]):
            seq.append(2)
        else:
            seq.append(3)
    return seq
```

### Question 4

```python
def Proba_Apparition(M,w,i0):
    """entrée: seq : séquence sur laquelle on cherche le mot positionné
               M: matrice de transition, w : le mot cherché , pos: position du mot dans la séquence
       sortie: la probabilité d'apparition du mot w à la position pos
    """
    if i0!= w[0]:
        return 0.0
    else:
        p=M[i0,w[1]]
        for i in range(2,len(w)):
            p*=M[w[i-1],w[i]]
        return p
        
```

### Question 5

```python
def calcStationnaire(M,f0):
    """
    entrées: M : la matrice de transition, f0: la fréquence d'apparition à l'état 0
    sortie: le vecteur des prbabilités stationnaires
    
    La méthode consiste à réaliser des exponentiations jusqu'à convergence et multiplier la matrice obtenu par f0
    """
    P = M
    while(True):
        tmp = M@P
        if np.allclose(P,tmp):
            break
        else:
            P = tmp
    return np.dot(P,f0)
freq_stationnaire = calcStationnaire(M,np.array(frequence))
print(frequence)
print(freq_stationnaire)
```

En ayant les probabilités stationnaires et les probabilités de transitions, le nombre attendu d'occurences pour un mot $w = a_1a_2\dots a_k$ de longeur $k$ est donnée par:
$$(l-k+1)\cdot P(w) = P(a_1)P(a_2|a_1)\dots P(a_k|a_{k-1}) $$

```python
def nbAttenduMarkov(M,freq_stationnaire,code,k,l):
    w = decode(code,k)
    p = freq_stationnaire[w[0]]
    for i in range(1,k):
        p*= M[w[i-1],w[i]]
    return (l-k+1)*p

def comptage_attendu_markov(M,freq_stationnaire,k,l):
    comptages = []
    for i in range(4**k):
        comptages.append(nbAttenduMarkov(M,freq_stationnaire,i,k,l) )
    return comptages
```

### Question 2
L'approximation qui peut être faite avec une loi de Poisson est la probabilité d'apparition du mot $w$ dans la séquence aléatoire de taille infini. Car la loi binomiale converge en loi vers la loi de Poisson.
### Question 3

```python
def distribution_poisson(code,k,M,freq_stationnaire,l):
    lamda=nbAttenduMarkov(M,freq_stationnaire,code,k,l)
    n=50
    occ_n=np.zeros(n)
    a=1
    occ_n[0]=1
    for i in range(1,n):
        a*=lamda
        a/=i
        occ_n[i]=a
    return occ_n 
    
```

```python
liste_mot = [ [0,0,0,0,0,0], [0,3,1,3,2,1], [0,3,0,3,0,3], [3,3,3,0,0,0]]
k = 6
l=1000
N = [i for i in range(50)]
x=['AAAAAA','ATCATCTGC','ATATAT','TTTAAA']
plt.figure(figsize=(10,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    occ_n = distribution_poisson(code(liste_mot[i],k),k,M,freq_stationnaire,l)
    plt.scatter(N,occ_n)
    plt.xlabel("n: le nombre d'apparitions")
    plt.ylabel("P(N>n)")
    plt.title("distribution du mot "+x[i])
```

En faisant l'hypothèse que la loi binomiale décrite ci-dessus converge vers une loi de Poisson de paramètre $\lambda = (l-k+1)\cdot P(w)$ qui s'agit du nombre attendu d'apparitions du mot $w$, on observe que les mots 'ATCATCTGC','ATATAT' on la même distribution. Cependant ceci n'est pas le cas pour 'ATATAT' et 'TTTAAA'. De plus, pour tous ces mots on observe que le nombre maximal d'apparition est presque surement inférieur à 10.

```python
def probas_empirique(code,k,M,f0,l,nbTests):
    occ_n = np.zeros(50)
    for i in range(nbTests):
        seq = simulationSequenceDinucleo(l,M,f0)
        occ = nbOccurencesMot(seq,k) #tableau trié par ordre lexicographique des mots de longueur k
        j = occ[code] 
        for a in range(0,j+1):
            occ_n[a]+=1
    occ_n/=nbTests
    return occ_n
```
